```python
import numpy as np
import kmeans
import common
import naive_em
import pandas as pd
import em

import matplotlib.pyplot as plt
plot_style = {
    # For latex-quality, i.e., publication quality legends and labels on graphs.
    "text.usetex": True,  # <----------------- you must have installed LaTeX on your system. Else set it to False
    
    # configure fonts
    "font.size":12,             # default font size
    "font.family":'serif',
    "legend.fontsize": 10,
    "axes.labelsize": 12,
    "axes.titlesize":14,
    "xtick.labelsize" : 10,
    "ytick.labelsize" : 10,
    
    # configure figure
    "figure.dpi": 144,
    "figure.figsize":'10,5',
    "figure.titlesize":14
}

plt.style.use ('ggplot')                      # emulate ggplot style
plt.rcParams.update(plot_style)
```

```python
X = np.loadtxt("toy_data.txt")
```

```python


for K in [1, 2, 3, 4]:
    print(f'K: {K}')
    kmeans_df = pd.DataFrame(columns=['mixture', 'post', 'cost'])
    em_df = pd.DataFrame(columns=['mixture', 'post', 'cost'])
    for seed in [0, 1, 2, 3, 4]:
        mixture, post = common.init(X, K, seed)
        kmeans_df.loc[seed] = kmeans.run(X, mixture, post)
        em_df.loc[seed] = naive_em.run(X, mixture, post)

    print(kmeans_df['cost'].min())
    print(em_df['cost'].min())
    
    m_km, p_km, c_km = kmeans_df.loc[kmeans_df['cost'].argmin()]
    m_em, p_em, c_em = em_df.loc[em_df['cost'].argmin()]
    
    common.plot(X, m_km, p_km, f'K = {K}, Kmeans algorithm')
    common.plot(X, m_em, p_em, f'K = {K}, EM algorithm')
```
